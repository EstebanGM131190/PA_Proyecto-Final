# Red Neuronal Retroalimentada Para predición de Precio de la Energia

Cargamos las librerias
Pandas para importar y trabajar con la hoja de datos
Matplotlib para ver los resultados de manera visual
Pyrenn Como nuestra libreria principal Encargada del manejo de la Red
Numpy para poder trabajar con nuestros datos de manera mas comoda.

```python
import pandas as pd
import matplotlib.pyplot as plt
import pyrenn as prn
import numpy as np

```

Cargamos los Datos de los precios de la energía (centavos por kilowatt/hora)

```python

#Obtención De los Datos (Precios de la electricidad del 2007 al 2017)
df = pd.ExcelFile('C:/Users/esteban/Documents/ElecPrice2.xls').parse('Hoja1')
```

Seleccionamos solamente la primera Hilera de datos la cual contiene el precio total
Ajustamos el Array para que la gráfica correspondiente sea correcta.
Seleccionamos los primeros 72 meses 2007 al 2011 como datos de entrenamiento
Seleccionamos el total de los datos 129 meses o 4 años y 9 meses como datos de prueba.

```python

Data = df.loc[0,:]      #Obtenemos la primera Hilera de los datos. (Precio total)

D = np.array(Data)      #Debido a que el primer dato que obtiene Pandas es el numero de la hilera tenemos que eliminarlo
D  = np.delete(D,0)
D2 = np.array(D[0:123])
#Seleccion de datos de entrenamiento y de Prueba
DataTrain= D[0:124]      #Datos de entrenamiento
DataTest = D[0:129]     #Datos de prueba
```

Creamos una red neuronal
Es una red de una entrada y una salida con dos capas intermedias cada una de 2 neuronas [1 2 2 1]
dIn = [0] Es decir no tiene delays en la entrada de la red.
DIntern = [1] Existe una retroalimentación de 1 periodo de tiempo entre las capas de la red
dOut = [1,2] Existe una retroalimentación de la capa de salida a la capa de entrada de 1 y 2 periodos de tiempo


```python

#%% Creamos una red neuronal
##Es una red de una entrada y una salida con tres capas intermedias cada una de 10 neuronas [1 10 10 10 1]
##dIn = [0] Es decir no tiene delays en la entrada de la red.
##DIntern = [1] Existe una retroalimentacion de 1 periodo de tiempo entre las capas de la red
##dOut = [1,2,3] Existe una retroalimentacion de la capa de salida a la capa de entrada de 1,2 y 3 periodos de tiempo
net = prn.CreateNN([1,10,10,10,1],dIn=[0],dIntern=[1],dOut=[1,2,3,4])
```

Entrenamos la red por medio del algoritmo Levenberg-Marquardt
net es la red
verbose = true Activa las salidas a consola
K-max es el máximo de iteraciones que va a realizar al entrenar la red
E_stop Detiene el entrenamiento cuando el error global es menor a 1e-3

```python

## Entrenamos la red por medio del algoritmo Levenberg-Marquardt
## net es la red
##verbose = true Activa las salidas a consola
##K-max es el maximo de iteraciones que va a realizar al entrenar la red
##E_stop Detiene el entrenamiento cuando el error global es menor a 1e-3
net = prn.train_LM(D2,DataTrain[1:124],net,verbose=True,k_max=200,E_stop=1e-3)
```

Primero obtenemos la salida de la red con los datos de entrenamiento
Despues  probamos la red agregandole al final del array el ultimo dato obtenido y volviendola a correr hasta obtener todos los datos de los meses faltantes

```python

## Da la salida de la red dados los datos de entrenamiento y de prueba
y = prn.NNOut(DataTrain,net)
#%%
    
ytest = prn.NNOut(D2,net)
ytest = np.append(D2,ytest[122])
#%%
for i in range(1, 7):
    ytest = prn.NNOut(ytest,net)
    ytest = np.append(D2,ytest[122:122+i])
```

Finalmente Graficamos los resultados.

```python

# Tamaño de la grafica
fig = plt.figure(figsize=(11,7))
# Creamos dos graficas
ax0 = fig.add_subplot(211)
ax1 = fig.add_subplot(212)
fs=18   #Tamaño del texto

#Train Data
ax0.set_title('Train Data',fontsize=fs)
ax0.plot(y,color='b',lw=2,label='NN Output')    #color azul grosor de la linea = 2
ax0.plot(DataTrain,color='r',marker='None',linestyle=':',lw=3,markersize=8,label='Train Data')
ax0.tick_params(labelsize=fs-2)
ax0.legend(fontsize=fs-2,loc='upper left')
ax0.grid()

#Test Data
ax1.set_title('Test Data',fontsize=fs)
ax1.plot(ytest,color='b',lw=2,label='NN Output')
ax1.plot(DataTest,color='r',marker='None',linestyle=':',lw=3,markersize=8,label='Test Data')
ax1.tick_params(labelsize=fs-2)
ax1.legend(fontsize=fs-2,loc='upper left')
ax1.grid()

fig.tight_layout()
plt.figure()
plt.show()
```

```python

Error_percentage = np.abs((DataTest[124:128]-ytest[124:128])/DataTest[124:128])
print(Error_percentage)
```
